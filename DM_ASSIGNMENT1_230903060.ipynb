```python
# Define a function to find the biggest correlation given the pd dataframe correlation matrix

def find_biggest_corr(df_corr):
  cols=df_corr.columns

  #initialize
  corr_max = 0
  i_max = 0
  j_max = 0

  #Find by exploring each element.
  for i in range(len(col)):
    for j in range(i+1,len(col)):
      corr_i_j = df_corr[cols[i]].iloc[j]
      if abs(corr_max) < abs(corr_i_j):
        corr_max = corr_i_j
        i_max = i
        j_max = j

  return corr_max,i_max,j_max
```

```python
# Find the pair of distinct numerical features with the highest correlation
df_corr = df.corr(method='pearson')

__, x, y = find_biggest_corr(df_corr)
print("The pair of distinct numerical features with the highest correlation is {} and {}.\n".format(df_corr.columns[x],df_corr.columns[y]))

# Plot
df.plot.scatter(df_corr.columns[x],df_corr.columns[y])
```

```python
# Use PCA and transfrom the dataframe.
from sklearn.decomposition import PCA

df_DropTarget = df.drop(['target'], axis=1, inplace=False) # do no use target in PCA
numComponents = 2
pca = PCA(n_components = numComponents)
pca.fit(df_DropTarget)

df_transform = pca.transform(df_DropTarget)
df_transform = pd.DataFrame(df_transform,columns=['pc1','pc2'],index=range(0,178))
df_transform['target'] = df['target']
display(df_transform)
```

```python
# Visualise
import matplotlib.pyplot as plt

colours = {0:'r', 1:'b', 2:'g'}

for col_target in colours:
    d = df_transform[df_transform['target']==col_target]
    plt.scatter(d['pc1'],d['pc2'],c=colours[col_target])

plt.xlabel("pc1")
plt.ylabel("pc2")
plt.legend(labels=['0','1','2'])
```

```python
print("pc1 contains {:.2%} of total message.".format(df_transform["pc1"].var()/sum(df_DropTarget.var())))
print("pc2 contains {:.2%} of total message.".format(df_transform["pc2"].var()/sum(df_DropTarget.var())))
```

I can obtain from the scatterplot that wines of target 0 can be relatively easy to be distinguished, but it is still hard to tell th difference between type 1 and type 2. This is possibly because these 2 components are still not good to separate the variables because pc1 contains 99.81% of the total message (variance), almost all the information. In this situation, standardisation should be performed before using PCA.

```python
# Standardisation
for col in df_DropTarget.columns:
  df_DropTarget[col] = (df_DropTarget[col] - df_DropTarget[col].mean())/df_DropTarget[col].std()
display(df_DropTarget)

# Then apply PCA
pca.fit(df_DropTarget)

df_transform = pca.transform(df_DropTarget)
df_transform = pd.DataFrame(df_transform,columns=['pc1','pc2'],index=range(0,178))
df_transform['target'] = df['target']
display(df_transform)
print("pc1 contains {:.2%} of total message.".format(df_transform["pc1"].var()/sum(df_DropTarget.var())))
print("pc2 contains {:.2%} of total message.".format(df_transform["pc2"].var()/sum(df_DropTarget.var())))

for col_target in colours:
    d = df_transform[df_transform['target']==col_target]
    plt.scatter(d['pc1'],d['pc2'],c=colours[col_target])

plt.xlabel("pc1")
plt.ylabel("pc2")
plt.legend(labels=['0','1','2'])
```

```python
# Create cubes
import cubes
from cubes import Workspace

engine = create_engine('sqlite:///data.sqlite')
create_table_from_csv(engine,
                      "country_income_adjusted.csv",
                      table_name="country_income",
                      fields=[
                          ("Region", "string"),
                          ("Age", "integer"),
                          ("Income", "integer"),
                          ("Online_Shopper", "string")],
                      create_id=True
                     )

workspace = Workspace()
workspace.register_default_store("sql", url="sqlite:///data.sqlite")
workspace.import_model("model.json")
cube = workspace.cube("country_income")
browser = workspace.browser(cube)
```

```python
with open("model.json", "r", encoding="utf-8") as f:
    model = f.read()
print(model)
```

b)

```python
# For the whole data cube
result = browser.aggregate()
print("For the whole data cube, the aggregate results are:")
print("Total income: {}".format(result.summary["total_income"]))
print("Average income: {}".format(result.summary["average_income"]))
print("Maximum income: {}".format(result.summary["maximum_income"]))
print("Minimum income: {}".format(result.summary["minimum_income"]))
print("\n")

# For the results per region
result = browser.aggregate(drilldown=["Region"])
for record in result:
    print(record)
print("\n")

# For the results per online activity
result = browser.aggregate(drilldown=["Online_Shopper"])
for record in result:
    print(record)
print("\n")

# For all people aged between 40 and 50
cuts = [cubes.RangeCut("Age", [40], [50])]
cell = cubes.Cell(cube, cuts)
result = browser.aggregate(cell)
print("The aggregate results for people aged from 40 to 50 is:")
print(result.summary)
```
